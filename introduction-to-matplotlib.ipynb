### Matplotlib example workflow

```python
# 0 . import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x=[1,2,3,4]
y=[11,22,33,44]

# 2. Setup Plot
fig,ax = plt.subplots(figsize=(10,10))

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save & Show (you save the whole fig)
fig.savefig("pltImages/sample-plot.png")
```

### Which one should you use ? (pyplot vs matplotlib OO method)

* When plotting something quickly, okay to use pyplot method
* When plotting something more advanced use OO method

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
# Pyplot Method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# OO Method mixed with pyplot
fix,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x='age',
            y='chol',
            c='target',
            ax=ax);
```

```python
# OO Method from scratch
fig,ax=plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50.age,
                    y=over_50.chol,
                    c=over_50.target);

# customize
ax.set(title="Heart Disease and Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholestrol");

# Add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--")
```

```python
over_50.head()
```

```python
# Subplots of chol , age , thalech
plt.style.use('fivethirtyeight')
fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10),
                          sharex=True);

# Add data to ax0
scatter_ax0 = ax0.scatter(x=over_50.age,
                     y=over_50.chol,
                     c=over_50.target);

# Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
       ylabel="Cholestrol");

# Adding a Legend to ax0
ax0.legend(*scatter_ax0.legend_elements(),title="Target");

# Add a mean line 
ax0.axhline(y=over_50.chol.mean(),
           linestyle='--')

## ax1
# Add data to ax1
scatter_ax1 = ax1.scatter(x=over_50.age,
                         y=over_50.thalach,
                         c=over_50.target);

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate");

# Adding a legend to ax1
ax1.legend(*scatter_ax1.legend_elements(),title="Target");

# Add a mean line
ax1.axhline(y=over_50.thalach.mean(),
           linestyle="--");

# Add a tile to fig
fig.suptitle("Heart Disease Analysis",fontsize=30,fontweight="bold");
```

```python
# Set The Style 
plt.style.use('default')


fig,ax=plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50.age,
                    y=over_50.chol,
                    c=over_50.target,
                    cmap="summer");

# customize
ax.set(title="Heart Disease and Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholestrol");

# Add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--",
          c='red');
```

```python
# Subplots of chol , age , thalech
plt.style.use('default')
fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10),
                          sharex=True);

# Add data to ax0
scatter_ax0 = ax0.scatter(x=over_50.age,
                     y=over_50.chol,
                     c=over_50.target,
                    cmap="winter");

# Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
       ylabel="Cholestrol");

# Adding a Legend to ax0
ax0.legend(*scatter_ax0.legend_elements(),title="Target");

# Add a mean line 
ax0.axhline(y=over_50.chol.mean(),
           linestyle='--')

ax0.set_xlim([50,80])

## ax1
# Add data to ax1
scatter_ax1 = ax1.scatter(x=over_50.age,
                         y=over_50.thalach,
                         c=over_50.target,
                         cmap="summer");

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate");

# Adding a legend to ax1
ax1.legend(*scatter_ax1.legend_elements(),title="Target");

# Add a mean line
ax1.axhline(y=over_50.thalach.mean(),
           linestyle="--");
ax1.set_xlim([50,80])

# Add a tile to fig
fig.suptitle("Heart Disease Analysis",fontsize=30,fontweight="bold");
```
